```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr

```

```python
sp500= pdr.get_data_yahoo('^gspc', start='1900-01-01')
sp500
```

```python
x= sp500['Close']
x.plot(grid=True)
```

```python
x.plot(grid=True, logy=True)
```

```python
lx= np.log10(x)
lx.plot(grid=True)
```

```python
lx= np.log2(x)
lx.plot(grid=True)
```

```python
t0= lx.index[0]
t1= lx.index[-1]
dt= t1-t0
dt
```

```python
dt.days
```

```python
dt.years= dt.days/365
dt.years
```

```python
dt.months= dt.years*12
dt.months
```

```python
dMoney= lx[-1]-lx[0]
dMoney
```

```python
dMoney= lx.max()-lx.min()
dMoney
```

```python
dMoney/dt.years
```

```python
lx.plot()
lxMean= lx.rolling(100, center=True).mean()
lxMean.plot(grid=True)
lxStd= lx.rolling(100, center=True).std()
lxStd.plot(grid=True)
```

```python
((lx-lxMean)/lxStd).plot()
```

```python
((lx-lxMean)/lxStd).hist(bins=100)
```
